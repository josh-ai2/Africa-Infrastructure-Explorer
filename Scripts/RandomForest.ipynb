# R7 Single decision tree

```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text

# Load the dataset from the specified filepath
file_path = '/Users/joshmanto/Downloads/KL/Main-new/Source/FILTERED DATASET/r7_filtered.csv'
df = pd.read_csv(file_path)

# Define the features and target variable
y = df['Class 1']  # Target variable: EA_SVC_B (Class 1)
X = df[['Class 2', 'Class 3', 'Class 4', 'Class 5', 'Class 6']]  # Features

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)

# Fit the model
decision_tree.fit(X_train, y_train)

# Export the decision tree rules as text
rules_r7 = export_text(decision_tree, feature_names=list(X.columns))

# Print the decision tree rules
print(rules_r7)
```

The code produces a single decision tree. It’s not an ensemble model like a Random Forest, where multiple trees are generated. It is unclear whether the tree is generalizable to represent the entire dataset, it is generated from a subset of the training data. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz

# Calculate Gini importance
gini_importance = decision_tree.feature_importances_

# Display Gini importance
for feature, importance in zip(X.columns, gini_importance):
    print(f"Feature: {feature}, Importance: {importance:.4f}")

# Visualize Gini importance
plt.figure(figsize=(10, 6))
plt.bar(X.columns, gini_importance, color='skyblue')
plt.title('Feature Importance (Gini)')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Print the depth of the tree
depth = decision_tree.get_depth()
print(f"Tree Depth: {depth}")

# Print the number of leaves in the tree
n_leaves = decision_tree.get_n_leaves()
print(f"Number of Leaves: {n_leaves}")

# Print the accuracy of the model
accuracy = decision_tree.score(X_test, y_test)
print(f"Accuracy: {accuracy:.4f}")

# Predict class probabilities for the first 5 test samples
probabilities = decision_tree.predict_proba(X_test[:5])
print("Predicted Class Probabilities for the first 5 test samples:")
print(probabilities)

# Print the actual feature values for the first 5 test samples
print("\nFeature values for the first 5 test samples:")
# Combine the features and predictions into a single DataFrame
results_df = X_test.iloc[:5].copy()
results_df['Prediction = 0'] = probabilities[:, 0] * 100  # Convert to percentage
results_df['Prediction = 1'] = probabilities[:, 1] * 100  # Convert to percentage

# Rename the index as 'ROW' to match the row numbers with the original data
results_df.index.name = 'ROW'

# Print the results
print(results_df.to_string(formatters={'Prediction = 0': '{:.2f}%'.format, 
                                       'Prediction = 1': '{:.2f}%'.format}))

# Export the decision tree to a Graphviz format and display
dot_data = export_graphviz(decision_tree, out_file=None, 
                           feature_names=X.columns,  
                           class_names=['0', '1'],  
                           filled=True, rounded=True,  
                           special_characters=True)  

# Create and display the graph using graphviz
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # Save the graph to a file (e.g., decision_tree.pdf)
graph.view()  # This will open the rendered tree visualization
```

# R7 Random Forest

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Load the dataset
file_path = '/Users/joshmanto/Downloads/KL/Main-new/Source/FILTERED DATASET/r7_filtered.csv'  # Replace with the actual path
df = pd.read_csv(file_path)

# Define the features and target variable
y = df['Class 1']  # Target variable: EA_SVC_B (Piped Water System)
X = df[['Class 2', 'Class 3', 'Class 4', 'Class 5', 'Class 6']]  # Features

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
n_estimators = 100
random_forest = RandomForestClassifier(n_estimators=n_estimators, random_state=42)

# Training the model with tqdm progress bar
print("Training the Random Forest model...")
for _ in tqdm(range(n_estimators)):
    random_forest.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = random_forest.score(X_test, y_test)
print(f"\nRandom Forest Accuracy: {accuracy:.4f}")

# Feature importance
importances = random_forest.feature_importances_
for feature, importance in zip(X.columns, importances):
    print(f"Feature: {feature}, Importance: {importance:.4f}")

# Visualize the feature importance
plt.figure(figsize=(10, 6))
plt.bar(X.columns, importances, color='skyblue')
plt.title('Feature Importance (Gini)')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Export the trained model to a file
model_filename = 'random_forest_model.pkl'
joblib.dump(random_forest, model_filename)
print(f"Model saved to {model_filename}")

# Load the model from the file (to verify that it works)
loaded_model = joblib.load(model_filename)

# Use the model to make predictions
predictions = loaded_model.predict(X_test)
print("Predictions for the first 5 test samples:")
print(predictions[:5])

# Optionally, predict probabilities for the first 5 samples
probabilities = loaded_model.predict_proba(X_test[:5])
print("Predicted Class Probabilities for the first 5 test samples:")
print(probabilities)

# Print the actual feature values for the first 5 test samples
print("\nFeature values for the first 5 test samples:")
results_df = X_test.iloc[:5].copy()
results_df['Prediction = 0'] = probabilities[:, 0] * 100  # Convert to percentage
results_df['Prediction = 1'] = probabilities[:, 1] * 100  # Convert to percentage

# Rename the index as 'ROW' to match the row numbers with the original data
results_df.index.name = 'ROW'

# Print the results
print(results_df.to_string(formatters={'Prediction = 0': '{:.2f}%'.format, 
                                       'Prediction = 1': '{:.2f}%'.format}))
```

In Random Forest, each tree is independently trained on a bootstrap sample of the data. The training process doesn't iterate over the entire dataset multiple times as it would in a neural network. Instead, the training is completed after all the trees are built.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Load the dataset
file_path = '/Users/joshmanto/Downloads/KL/Main-new/Source/FILTERED DATASET/Added Variables filtered Dataset/r7-filtered-extravar.csv'  # Replace with the actual path
df = pd.read_csv(file_path)

# Handle any missing values
df = df.fillna(-1)  # Replace missing values with -1 or choose another method if appropriate

# Define the features and target variable
# Predicting EA_SVC_B using other variables except EA_GPS_LA and EA_GPS_LO
y = df['EA_SVC_B']  # Target variable
X = df[['EA_SVC_A', 'EA_SVC_C', 'Q56I', 'Q8B', 'Q92B', 'Q92A', 'Q93', 'Q56M', 
        'EA_FAC_D', 'EA_FAC_C', 'EA_FAC_B']]  # Features excluding EA_GPS_LA and EA_GPS_LO

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
n_estimators = 100
random_forest = RandomForestClassifier(n_estimators=n_estimators, random_state=42)

# Training the model with tqdm progress bar
print("Training the Random Forest model...")
for _ in tqdm(range(n_estimators)):
    random_forest.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = random_forest.score(X_test, y_test)
print(f"\nRandom Forest Accuracy: {accuracy:.4f}")

# Feature importance
importances = random_forest.feature_importances_
for feature, importance in zip(X.columns, importances):
    print(f"Feature: {feature}, Importance: {importance:.4f}")

# Visualize the feature importance
plt.figure(figsize=(10, 6))
plt.bar(X.columns, importances, color='skyblue')
plt.title('Feature Importance (Gini)')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Export the trained model to a file
model_filename = 'random_forest_model_round8.pkl'
joblib.dump(random_forest, model_filename)
print(f"Model saved to {model_filename}")

# Load the model from the file (to verify that it works)
loaded_model = joblib.load(model_filename)

# Use the model to make predictions
predictions = loaded_model.predict(X_test)
print("Predictions for the first 5 test samples:")
print(predictions[:5])

# Optionally, predict probabilities for the first 5 samples
probabilities = loaded_model.predict_proba(X_test[:5])
print("Predicted Class Probabilities for the first 5 test samples:")
print(probabilities)

# Print the actual feature values for the first 5 test samples
print("\nFeature values for the first 5 test samples:")
results_df = X_test.iloc[:5].copy()
results_df['Prediction = 0'] = probabilities[:, 0] * 100  # Convert to percentage
results_df['Prediction = 1'] = probabilities[:, 1] * 100  # Convert to percentage

# Rename the index as 'ROW' to match the row numbers with the original data
results_df.index.name = 'ROW'

# Print the results
print(results_df.to_string(formatters={'Prediction = 0': '{:.2f}%'.format, 
                                       'Prediction = 1': '{:.2f}%'.format}))
```
